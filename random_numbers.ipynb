
**Write a program that generates 100 random integers between 0 and 9**
1. Print them in a 10 by 10 matrix neatly arranged (one space between each number)
2. If the random number is an odd number, print "@" instead
3. Calculate and Print the total of each row like the following (use "*" to separate the total from the numbers)

**Sample output for each of the above is shown below**

```python
from random import randint
```

```python
# Printing the 10*10 int matrix
for i in range(10):
    for i in range(10):
        print (randint(0, 9),end=" ")
    print('\n')
```

```python
# printing the matrix with @ at odd numbers place
for i in range(10):
    for i in range(10):
        num=randint(0, 9)
        if num%2 != 0:
            print ('@',end=" ")
        else:
            print(num, end=" ")
    print('\n')
```

```python
#printing the matrix with sum
for i in range(10):
    row_sum=0
    for i in range(10):
        num=randint(0, 9)
        row_sum= row_sum + num
        print(num, end=" ")
    print('*',row_sum)
    print('\n')
```
